# Merging the data together

```python
# appearances and players 
# df - appearances, df1 = game_events, df2 = game_lineups, df3 = games, df4 = players. 

app_merge = pd.merge(df,df4, on = "player_id", how = "left")
app_merge.reset_index(drop = True, inplace = True)
app_merge
```

```python
# merging game events using the "game_id" which is in both columns 
# # df - appearances, df1 = game_events, df2 = game_lineups, df3 = games, df4 = players. app_merge = appearances and players

events_merge = pd.merge(app_merge, df1, on = "game_id", how = "left")
events_merge.reset_index(inplace = True)
events_merge
```

```python
# merging game lineups with its common column 
# # df - appearances, df1 = game_events, df2 = game_lineups, df3 = games, df4 = players.
# app_merge = appearances and players, events_merge = appearances, players and game_events 

lineups_merge = pd.merge(app_merge, df2, on = "game_id", how = "left")
lineups_merge.reset_index(drop = True, inplace = True)
lineups_merge
```

```python
# Merging the games column 
# df - appearances, df1 = game_events, df2 = game_lineups, df3 = games, df4 = players.
# app_merge = appearances and players, events_merge = appearances, players and game_events and 
# lineups_merge = appearances, players, game_event and game_lineups

Final_data = pd.merge(lineups_merge, df3, on = "game_id", how = "left")
Final_data.reset_index(drop = True, inplace = True)
Final_data
```
